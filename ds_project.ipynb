```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('C:\\Users\\paint\\Downloads\\Sample_Data.csv')
df
```

```python
df[['Date', 'Time']] = df['Date_Time'].str.split(' ', expand=True)
df
```

```python
df.groupby(['Month','Year']).agg({'total_count':'mean'})
```

```python
sns.lineplot(x="Month" , y="total_count",data=df,hue="Year")
```

```python
sns.barplot(x="Month" , y="total_count",data=df,hue="Year")
```

```python
df.groupby(['Month','Year',"is_holiday"]).agg({'total_count':'mean'})
```

```python
sns.catplot(x="Month" , y="total_count",data=df,hue="Year",kind="bar" , height=10 , errorbar=('ci', False),palette="crest")
```

```python
time1=df[(df["Time"]>="00:00") & (df["Time"]<"06:00")]
time1
```

```python
time2=df[(df["Time"]>="06:00") & (df["Time"]<"12:00")]
time2
```

```python
time3=df[(df["Time"]>="12:00") & (df["Time"]<"18:00")]
time3
```

```python
time4=df[(df["Time"]>="18:00") & (df["Time"]<"24:00")]
time4
```

```python
time1.shape[0]+time2.shape[0]+time3.shape[0]+time4.shape[0]
```

```python
sns.lineplot(x="Month" , y="total_count" , data=time1,label="00:00 to 06:00")
sns.lineplot(x="Month" , y="total_count" , data=time2 , label="06:00 to 12:00")
sns.lineplot(x="Month" , y="total_count" , data=time3 ,label="12:00 to 18:00")
sns.lineplot(x="Month" , y="total_count" , data=time4 , label="18:00 to 24:00")
plt.legend()
```

```python
sns.relplot(x="Month" , y= "total_count" , data=df , kind="line" , col="Time", hue="Year" ,col_wrap=2)
```

```python
sns.catplot(x="Month" , y= "total_count" , data=df , kind="bar" , col="Time", hue="Year" ,height=5,aspect=2)

sns.set_style("darkgrid")
plt.show()
```

```python
sns.barplot(x="Month" , y="total_count" , data=time1, edgecolor="black",alpha=0.2)
sns.barplot(x="Month" , y="total_count" , data=time2 ,edgecolor="black",alpha=0.2)
sns.barplot(x="Month" , y="total_count" , data=time3 ,edgecolor="black",alpha=0.2)
sns.barplot(x="Month" , y="total_count" , data=time4 ,edgecolor="black",alpha=0.2)
```

```python
sns.catplot(x="Month" , y= "total_count" , data=df , kind="bar" , col="is_holiday, hue="Year",height=5,aspect=2)
```

```python
sns.catplot(x="Time" , y= "total_count" , data=df , kind="bar" , col="Month", hue="Year",col_wrap=1 ,height=5,aspect=2,errorbar=('ci', False),palette="crest")
sns.set_style("darkgrid")
plt.show()
```
